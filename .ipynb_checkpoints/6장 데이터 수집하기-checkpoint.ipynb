### 덴마 수집

```python
# -*- coding: utf-8 -*-
from bs4 import BeautifulSoup
import requests

pre_pg = 0
cur_pg = 1

hrefs = []

while True:
    html = requests.get('http://comic.naver.com/webtoon/list.nhn?titleId=119874&weekday=tue&page={0}'.format(cur_pg)).text
    soup = BeautifulSoup(html, "html5lib")

    titles = soup.find_all("td", attrs={'class': 'title'})
    
    # 이전 페이지와 이번에 불러온 페이지가 동일하면
    if pre_pg == int(soup.find_all("span", attrs={'class': 'current'})[0].text.strip()): break
    
    pre_pg = cur_pg
    cur_pg = cur_pg + 1 
    
    hrefs = hrefs + ['[{0}] : http://comic.naver.com{1}'.format(title.text.encode('utf-8').strip(), title.find("a")['href'].encode('utf-8')) for title in titles]

for urls in hrefs:
    print urls
```
