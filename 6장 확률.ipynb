```python
import random
from __future__ import division   # 정확한 소수점 계산을 위해 import
from collections import Counter 
```

```python
def random_kid():
    return random.choice(["boy", "girl"])
```

```python
both_girls = 0
older_girl = 0
either_girl = 0

random.seed(0)

for _ in range(10000):
    younger = random_kid()
    older = random_kid()
    
    if older == "girl": # 첫째 아이가 딸
        older_girl += 1
    if older == "girl" and younger == "girl": # 첫째 둘째 모두 딸
        both_girls += 1
    if older == "girl" or younger == "girl": # 첫째가 딸이거나 둘째가 딸
        either_girl += 1

print "P(both|older):", both_girls / older_girl
print "P(both|either):", both_girls / either_girl
```

### 정규분포
* 정규분포는 종형 곡선 모양의 분포이며, 평균인 m(뮤)와 표준편차 시그마 의 두 파라미터로 정의된다.
* 표준정규분포는 m(뮤) = 0 이고 시그마 = 1 인 정규분포 

```python
import math
from matplotlib import pyplot as plt

%matplotlib inline

def normal_pdf(x, mu=0, sigma=1):
    sqrt_two_pi = math.sqrt(2*math.pi)
    return (math.exp(-(-x-mu) ** 2 / 2 / sigma ** 2) / (sqrt_two_pi * sigma))
```

```python
xs = [x / 10.0 for x in range(-50, 50)]
```

```python
plt.plot(xs, [normal_pdf(x,sigma=1) for x in xs], '-', label='mu=0, sigma=1')
plt.plot(xs, [normal_pdf(x,sigma=2) for x in xs], '-', label='mu=0, sigma=2')
plt.plot(xs, [normal_pdf(x,sigma=0.5) for x in xs], '-', label='mu=0, sigma=0.5')
plt.plot(xs, [normal_pdf(x,mu=-1) for x in xs], '-', label='mu=1, sigma=1')
plt.legend()
plt.title("Various Normal pdfs")
plt.show()
```

### 누적분포
* 정규분포의 누적분포함수를 간편하게 구현하기는 힘들 수도 있지만, 파이썬의 math.erf 를 사용하면 구현할 수 있다.

```python
def normal_cdf(x, mu=0, sigma=1):
    return (1 + math.erf((x-mu)/math.sqrt(2) / sigma)) / 2
```

```python
xs = [x / 10.0 for x in range(-50, 50)]
plt.plot(xs, [normal_cdf(x, sigma=1) for x in xs], '-', label='mu=0, sigma=1')
plt.plot(xs, [normal_cdf(x, sigma=2) for x in xs], '--', label='mu=0, sigma=2')
plt.plot(xs, [normal_cdf(x, sigma=0.5) for x in xs], ':', label='mu=0, sigma=0.5')
plt.plot(xs, [normal_cdf(x, mu=-1) for x in xs], '-.', label='mu=-1, sigma=1')

plt.legend(loc=4) #우측하단
plt.title("Various Normal cdfs")
plt.show()
```

### 중심극한정리
* 중심극한정리란 동일한 분포에 대한 독립적인 확률변수의 평균을 나타내는 확률변수가 대략적으로 정규분포를 따른다는 정리이다.

```python
def bernoulli_trial(p):
    return 1 if random.random() < p else 0

def binomial(n, p):
    return sum(bernoulli_trial(p) for _ in range(n))
```

```python
def make_hist(p, n, num_points):
    data = [binomial(n, p) for _ in range(num_points)]
    
    histogram = Counter(data)
    plt.bar([x-0.4 for x in histogram.keys()],
            [v / num_points for v in histogram.values()],
            0.8,
            color='0.75')
    mu = p * n
    sigma = math.sqrt(n * p * (1-p))
    
    #근사된 정규분포를 라인 차트로 표현
    xs = range(min(data), max(data) + 1)
    ys = [normal_cdf(i + 0.5, mu, sigma) - normal_cdf(i - 0.5, mu, sigma) for i in xs]
    
    plt.plot(xs, ys)
    plt.title("Binomial Distribution vs. Normal Approximation")
    plt.show()
```

```python
make_hist(0.75, 100, 10000)
```
